```python

from game import Game
from RL_Agent import RL_Agent

env = Game()
num_episodes = 10
```

```python

def test_agent():

    winning_rates = env.get_players_insights( num_episodes)
    player_total_won_policy = winning_rates[3]
    best_player_in_total = env.best_total_won_player(winning_rates[1])[0].strategy.strategy_factors()
    #agent = RL_Agent(best_player_in_total,  player_total_won_policy)
    winning_rate_of_policy = winning_rates[1][env.best_total_won_player(winning_rates[1])[0].name]

    for episode in range(num_episodes):
        agent = RL_Agent(best_player_in_total,  player_total_won_policy)
        chosen_action = agent.select_action()
        reward = 0

        if winning_rate_of_policy > 45:
            reward = 1
    
        RL_policy = agent.update_q_values_plus(chosen_action, reward)
        policy_name = "RL_plus_policy"

        env.evaluate_policy(RL_policy , policy_name , num_episodes)
        
    for episode in range(num_episodes):
        agent = RL_Agent(best_player_in_total,  player_total_won_policy)
        chosen_action = agent.select_action()
        reward = 0

        if winning_rate_of_policy > 45:
            reward = 1
    
        RL_policy =agent.update_q_values_minus(chosen_action, reward)
        policy_name = "RL_minus_policy"

        env.evaluate_policy(RL_policy , policy_name , num_episodes)
        
    print("Number of tests:" ,num_episodes *2 , "number of games inside a test:" , env.num_games  , " ,all games:" ,2*env.num_games* num_episodes)
    env.get_final_result()

test_agent()
```

```python
best_strategies = env.get_best_strategies()

collaborative_insights = env.evaluate_policy(best_strategies[0] , "Collaborative")
min_regret_insights = env.evaluate_policy(best_strategies[1]  , "Min_regret")
modified_collaborative_insights =env.evaluate_policy(best_strategies[2]  , "Mod_col")

env.visualize(collaborative_insights[1] , "Collaborative" , collaborative_insights[2])
env.visualize(min_regret_insights[1] , "Min_regret" , min_regret_insights[2])
env.visualize(modified_collaborative_insights[1] , "Mod_col" , modified_collaborative_insights[2])


print("Number of tests:" ,env.test_times , "number of games inside a test:" ,env.num_games )
```
